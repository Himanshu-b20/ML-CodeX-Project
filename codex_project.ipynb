```python
mapping = {
    "Metor": "Metro",
    "urbna": "Urban"
}

df_new['zone'] = df_new['zone'].replace(mapping)
df_new['zone'].unique()
```

```python
df_new['current_brand'].unique()
```

```python
mapping_current_brand = {
    "newcomer": "Newcomer",
    "Establishd": "Established"
}

df_new['current_brand'] = df_new['current_brand'].replace(mapping_current_brand)
df_new['current_brand'].unique()
```

```python
def age_cat(age):
    if age >=18 and age <=25:
        return '18-25'
    elif age >=26 and age <=35:
        return '26-35'
    elif age >=36 and age <=45:
        return '36-45'
    elif age >=46 and age <=55:
        return '46-55'
    elif age >=56 and age <=70:
        return '56-70'
    else:
        return '70+'
```

```python
df_new['age_group'] = df_new['age'].apply(age_cat)
df_new.head()
```

```python
df_new.drop('age', axis=1)
df_new.head()
```

```python
df_new['consume_frequency(weekly)'] = df_new['consume_frequency(weekly)'].map({
    '0-2 times' : 1,
    '3-4 times' : 2,
    '5-7 times' : 3
})
df_new.head()
```

```python
df_new['awareness_of_other_brands'].unique()
```

```python
df_new['awareness_of_other_brands'] = df_new['awareness_of_other_brands'].map({
    '0 to 1' : 1,
    '2 to 4' : 2,
    'above 4' : 3
})
df_new.head()
```

```python
df_new['cf_ab_score'] = round(df_new['consume_frequency(weekly)'] / (df_new['consume_frequency(weekly)'] + df_new['awareness_of_other_brands']), 2)
df_new.head()
```

```python
df_new['cf_ab_score'].max()
```

```python
df_new['zone'] = df_new['zone'].map({
    'Rural' : 1,
    'Semi-Urban' : 2,
    'Urban' : 3,
    'Metro' : 4,
})
df_new.head()
```

```python
df_new['income_levels'].unique()
```

```python
df_new['income_levels'] = df_new['income_levels'].map({
    'Not Reported' : 0,
    '<10L' : 1,
    '10L - 15L' : 2,
    '16L - 25L' : 3,
    '26L - 35L' : 4,
    '> 35L' : 5
})
df_new.head()
```

```python
df_new['zas_score'] = df_new['zone'] * df_new['income_levels']
df_new.head()
```

```python
from sklearn.model_selection import train_test_split
X = df_final.drop(['respondent_id', 'age', 'price_range'], axis=1)
X.head()
```

```python
y = df_final['price_range']
y.head()
```

```python
y=y.map({
    '50-100' : 0,
    '100-150' : 1,
    '150-200' : 2,
    '200-250' : 3
})
y.head()
```

```python
X['age_group'] = X['age_group'].map({
    '18-25' : 0,
    '26-35' : 1,
    '36-45' : 2,
    '46-55' : 3,
    '56-70' : 4
})
X.head()
```

```python
X['health_concerns'] = X['health_concerns'].map({
    'Low (Not very concerned)' : 0,
    'Medium (Moderately health-conscious)' : 1,
    'High (Very health-conscious)' : 2
})
X.head()
```

```python
X['preferable_consumption_size'] = X['preferable_consumption_size'].map({
    'Small (250 ml)' : 0,
    'Medium (500 ml)' : 1,
    'Large (1 L)' : 2
})
X.head()
```

```python
X_encoded = pd.get_dummies(X, columns=['gender', 'occupation', 'current_brand', 'reasons_for_choosing_brands', 'flavor_preference', 'purchase_channel', 'packaging_preference', 'typical_consumption_situations'], drop_first=True)
X_encoded.shape
```

```python
X_encoded.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.25, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
# from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
model_nb = GaussianNB()
model_nb.fit(X_train, y_train)

y_pred_nb = model_nb.predict(X_test)

accuracy = accuracy_score(y_test, y_pred_nb)
print("Accuracy for GaussianNB:", accuracy)
report = classification_report(y_test, y_pred_nb)
report_dict = classification_report(y_test, y_pred_nb, output_dict=True)
print(report)
```

```python
import mlflow
mlflow.set_experiment("CodeX Project Models")
mlflow.set_tracking_uri("http://127.0.0.1:5000/")

with mlflow.start_run(run_name="Gaussian Naive Bayes"):
    mlflow.log_params({'model_name' : 'GaussianNB'})
    mlflow.log_metrics({
        'accuracy': accuracy,
        'f1_score_macro': report_dict['macro avg']['f1-score']
    })
    mlflow.sklearn.log_model(model_nb, "GaussianNB")
```

```python
model_lr = LogisticRegression(max_iter=200)
model_lr.fit(X_train, y_train)

y_pred_lr = model_lr.predict(X_test)

accuracy = accuracy_score(y_test, y_pred_lr)
print("Accuracy for LogisticRegression: ", accuracy)
report = classification_report(y_test, y_pred_lr)
report_dict = classification_report(y_test, y_pred_lr, output_dict=True)
print(report)
```

```python
with mlflow.start_run(run_name="Logistic Regression"):
    mlflow.log_params({'max_iter' : 200})
    mlflow.log_metrics({
        'accuracy': accuracy,
        'f1_score_macro': report_dict['macro avg']['f1-score']
    })
    mlflow.sklearn.log_model(model_lr, "LogisticRegression")
```

```python
model_svc = SVC(kernel='rbf', C=1.0, gamma='scale')
model_svc.fit(X_train, y_train)

y_pred_svc = model_svc.predict(X_test)

accuracy = accuracy_score(y_test, y_pred_svc)
print("Accuracy for SVC: ", accuracy)
report = classification_report(y_test, y_pred_svc)
report_dict = classification_report(y_test, y_pred_svc, output_dict=True)
print(report)
```

```python
with mlflow.start_run(run_name="SVC"):
    mlflow.log_params({'kernel' : 'rbf', 'C':1.0, 'gamma': 'scale'})
    mlflow.log_metrics({
        'accuracy': accuracy,
        'f1_score_macro': report_dict['macro avg']['f1-score']
    })
    mlflow.sklearn.log_model(model_svc, "SVC")
```

```python
model_rfc = RandomForestClassifier(n_estimators=100, random_state=42)
model_rfc.fit(X_train, y_train)

y_pred_rfc = model_rfc.predict(X_test)

accuracy = accuracy_score(y_test, y_pred_rfc)
print("Accuracy for RFC: ", accuracy)
report = classification_report(y_test, y_pred_rfc)
report_dict = classification_report(y_test, y_pred_rfc, output_dict=True)
print(report)
```

```python
with mlflow.start_run(run_name="Random Forest"):
    mlflow.log_params({'n_estimators' : 100, 'random_state':42})
    mlflow.log_metrics({
        'accuracy': accuracy,
        'f1_score_macro': report_dict['macro avg']['f1-score']
    })
    mlflow.sklearn.log_model(model_rfc, "RandomForestClassifier")
```

```python
model_xgb = XGBClassifier(eval_metric='logloss')
model_xgb.fit(X_train, y_train)

y_pred_xgb = model_xgb.predict(X_test)

accuracy = accuracy_score(y_test, y_pred_xgb)
print("Accuracy for RFC: ", accuracy)
report = classification_report(y_test, y_pred_xgb)
report_dict = classification_report(y_test, y_pred_xgb, output_dict=True)
print(report)
```

```python
with mlflow.start_run(run_name="XG Boost"):
    mlflow.log_params({'eval_metric' : 'logloss'})
    mlflow.log_metrics({
        'accuracy': accuracy,
        'f1_score_macro': report_dict['macro avg']['f1-score']
    })
    mlflow.sklearn.log_model(model_xgb, "XGBClassifier")
```
